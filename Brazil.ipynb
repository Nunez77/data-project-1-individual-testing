```python
import pandas as pd
import numpy as np
```

```python
my_content = pd.read_csv('Brazil_FIX_tracks_with_audio_features.csv')
```

```python
my_content.head()
```

```python
total_sum = my_content.groupby(['Year','Month']).Streams.sum()
```

```python
data = pd.DataFrame(total_sum)
```

```python
my_data = pd.DataFrame(data.to_records())
```

```python
my_data.head()
```

```python
final_data = my_content.merge(my_data, how='inner', left_on=['Year', 'Month'], right_on=['Year', 'Month'])
```

```python
final_data.head()
```

```python
final_data['Danceability %'] = final_data['Danceability']*final_data['Streams_x']/final_data['Streams_y']
```

```python
final_data['Energy %'] = final_data['Energy']*final_data['Streams_x']/final_data['Streams_y']
```

```python
final_data['Valence %'] = final_data['Valence']*final_data['Streams_x']/final_data['Streams_y']
```
